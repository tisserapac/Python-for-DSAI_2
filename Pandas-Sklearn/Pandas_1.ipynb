```python
#Exercise 2: Get the list of countries schooling more than the median
#and GDP more than the median
#df['Schooling']
median_schooling = df['Schooling'].median()
schooling_gt_median = df['Schooling'] > median_schooling
median_gdp = df['GDP'].median()
gdp_gt_median = df['GDP'] > median_gdp
schooling_gt_median_and_gdp_gt_median = schooling_gt_median & gdp_gt_median
#schooling_gt_median_and_gdp_gt_median.unique()
print(f"Number of countries with schooling more than the median and GDP more than the median is : ", len(df['Country'][schooling_gt_median_and_gdp_gt_median].unique()))

```

```python
# Professor's Answer

# () function - can put params
# no bracket - attribute

cond1 = df.Schooling > df.Schooling.median() 
cond2 = df.GDP > df.GDP.median() 

contries = df.Country[cond1 & cond2].unique()
type(contries)
#print(f"Countries with schooling more than the median and GDP more than the median is : ", contries)
```
